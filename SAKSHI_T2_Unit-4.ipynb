```python
import pandas as pd
df=pd.read_csv("polylinearregression.csv")
df
```

```python
x=df.iloc[:,1:2]
y=df.iloc[:,-1]
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
ypred=lr.predict(x)
print(ypred)
```

```python
from sklearn import metrics
print("MAE:",metrics.mean_absolute_error(y,ypred))
print("MSE:",metrics.mean_squared_error(y,ypred))
print("R2 score:",metrics.r2_score(y,ypred))
```

```python
import seaborn as sns
sns.regplot(data=df,x='Temperature',y='Pressure')
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=3)
x_poly=poly.fit_transform(x)
print(x_poly)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_poly,y)
ypred=lr.predict(x_poly)
print(ypred)
```
