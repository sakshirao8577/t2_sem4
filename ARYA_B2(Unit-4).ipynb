# Linear Regression
- Statistical Method used to model the relationship between a dependent variable and one or more independent variable
- The goal of LR is to find the best fitting line that describes the relationship between the variables.

```python
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
```

```python
print(type(x))
print(type(y))
```

```python
print(x)
```

```python
print(y)
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=1)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
print(x_test)
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
ypred=lr.predict(x_test)
print(ypred)
```

```python
diff=pd.DataFrame({'Actual':y_test,'Predicted':ypred})
diff
```

```python
print("coefficient:",lr.coef_)
print("Intercept:",lr.intercept_)
```
